# $因子描述：过去N日夏普（Risk-Adjusted Momentum）$

***计算方式：过去N天收益/过去N天收益标准差。***  

![示例](sample.jpg)

## 计算

```python
# 参数平滑
sts_ad11 = (xy.rets.T.rolling(11).mean() / xy.rets.T.rolling(13).std()).T
sts_ad12 = (xy.rets.T.rolling(12).mean() / xy.rets.T.rolling(13).std()).T
sts_ad13 = (xy.rets.T.rolling(13).mean() / xy.rets.T.rolling(13).std()).T
sts_ad14 = (xy.rets.T.rolling(14).mean() / xy.rets.T.rolling(13).std()).T
sts_ad15 = (xy.rets.T.rolling(15).mean() / xy.rets.T.rolling(13).std()).T

_sts_ad_mean = (sts_ad11+sts_ad12+sts_ad13+sts_ad14+sts_ad15)/5
sts_ad_mean = xy.read_Factor(xy.trd_dates,'sts_ad', df=_sts_ad_mean.dropna(axis=1,how='all'))
```

## 评估

### cl-cl

```python
# 分组收益率
gr_ad = xy.cal_rate_group(
    sts_ad_mean.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    group_num=4,
    masks=1
)
gr_ad.cumsum().plot()
```

```python
# 单因子强弱
ad_select, ad_rets = xy.single_Factor_backtest(
    sts_ad_mean.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    _side=1,
    fee=0.000,
    name='ad'
)
```

```python
# 因子截面IC趋势
rankIC = xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False).corrwith(sts_ad_mean.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False))
rankIC.rolling(60).mean().plot(figsize=(15,8),grid=1)
rankIC.mean()
```

### cl_op

```python
# 分组收益率
gr_ad = xy.cal_rate_group(
    sts_ad_mean.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    group_num=4,
    masks=1
)
gr_ad.cumsum().plot()
```

```python
# 单因子强弱
ad_select, ad_rets = xy.single_Factor_backtest(
    sts_ad_mean.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    _side=1,
    fee=0.000,
    name='ad'
)
```

```python
# 因子截面IC趋势
rankIC = xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False).corrwith(sts_ad_mean.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False))
rankIC.rolling(60).mean().plot(figsize=(15,8),grid=1)
rankIC.mean()
```
