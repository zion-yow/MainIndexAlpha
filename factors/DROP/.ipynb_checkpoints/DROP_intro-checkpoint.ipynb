## 计算

```python
def find_drop_most(_arr):
    droping1 = (_arr[1:]/_arr[:-1] - 1)[-120:-5][np.where((abs((_arr[1:]/_arr[:-1] - 1)[-120:-5])>0.0001))]
    droping2 = (_arr[1:]/_arr[:-1] - 1)[:30][np.where((abs((_arr[1:]/_arr[:-1] - 1)[:30])>0.0001))]
    
    
    ss1 = np.nansum(droping1)/np.nanstd(droping1)
    ss2 = np.nansum(droping2)/np.nanstd(droping2)
    
    ss = ss1 
    return ss
```

```python
def cal_DROP(arr):
    _Drop = np.apply_along_axis(find_drop_most,1,xy.rolling_window(arr,240)[::240])
    Drop = pd.Series(_Drop,index = arr.index[::240])
    
    Drop.index = Drop.index.map(lambda x: x[:10])
    return Drop
```
