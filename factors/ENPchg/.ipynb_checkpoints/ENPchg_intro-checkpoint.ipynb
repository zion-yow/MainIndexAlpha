***当日一致预期净利润（东财）同比增速变化率***  

## 计算

```python
files = os.listdir(r'E:\PyProject\FPY1\Data\ENP')

files_lst = []
for f in files:
    files_lst.append(pd.read_csv(f'../../FPY1/Data/ENP/{f}',index_col=0).dropna(axis=1, how='all'))
st_FPY1 = pd.concat(files_lst,axis=1)
```

```python
NP_files =  os.listdir(r'E:\PyProject\FPY1\Data\NP')

NP_files_lst = []
for f in NP_files:
    NP_files_lst.append(pd.read_csv(f'../../FPY1/Data/NP/{f}',index_col=0).dropna(axis=1, how='all').iloc[1:,:].astype('float32'))
st_NP = pd.concat(NP_files_lst,axis=1)
st_NP = st_NP.mask(st_NP==0.0000000,np.nan).sort_index()
st_NP.columns = [str(d)[:10] for d in pd.to_datetime(st_NP.columns)]
st_NP = st_NP[~st_NP.index.duplicated()]
```

```python
NP_TTM_files =  os.listdir(r'E:\PyProject\FPY1\Data\NP_TTM')

NP_TTM_files_lst = []
for f in NP_TTM_files:
    NP_TTM_files_lst.append(pd.read_csv(f'../../FPY1/Data/NP_TTM/{f}',index_col=0).astype('float32'))
    # if str(f[-8:-4]) == '2021':
    #     break
st_NP_TTM = pd.concat(NP_TTM_files_lst,axis=1)
st_NP_TTM = st_NP_TTM.mask(st_NP_TTM==0.0000000,np.nan).sort_index()    
st_NP_TTM.columns = [str(d)[:10] for d in pd.to_datetime(st_NP_TTM.columns)]
st_NP_TTM = st_NP_TTM[~st_NP_TTM.index.duplicated()]
```

```python
ENP_files =  os.listdir(r'E:\PyProject\FPY1\Data\ENP')

ENP_files_lst = []
for f in ENP_files:
    ENP_files_lst.append(pd.read_csv(f'../../FPY1/Data/ENP/{f}',index_col=0).astype('float32'))
st_ENP = pd.concat(ENP_files_lst,axis=1)
st_ENP = st_ENP.mask(st_ENP==0.0000000,np.nan).sort_index()
st_ENP.columns = [str(d)[:10] for d in pd.to_datetime(st_ENP.columns)]
delist_st = set(st_NP.index) - set(st_ENP.index)
delist_ENP_st = pd.DataFrame(np.nan, index=list(delist_st), columns=st_NP.columns)
ENP_st = pd.concat([st_ENP,delist_ENP_st]).sort_index()
```

```python
pre_raise50 = ((st_ENP.mask(pd.isna(st_ENP),st_NP_TTM) * (xy.weights50)).loc[:,'2017-01-03':].dropna(how='all').mean() / (st_NP * (xy.weights50)).loc[:,'2017-01-03':].dropna(how='all').mean()-1).rename('上证50').dropna()
pre_raise300 = ((st_ENP.mask(pd.isna(st_ENP),st_NP_TTM) * (xy.weights300)).loc[:,'2017-01-03':].dropna(how='all').mean() / (st_NP * (xy.weights300)).loc[:,'2017-01-03':].dropna(how='all').mean()-1).rename('沪深300').dropna()
pre_raise500 = ((st_ENP.mask(pd.isna(st_ENP),st_NP_TTM) * (xy.weights500)).loc[:,'2017-01-03':].dropna(how='all').mean() / (st_NP * (xy.weights500)).loc[:,'2017-01-03':].dropna(how='all').mean()-1).rename('中证500').dropna()
pre_raise1000 = ((st_ENP.mask(pd.isna(st_ENP),st_NP_TTM) * (xy.weights1000)).loc[:,'2017-01-03':].dropna(how='all').mean() / (st_NP * (xy.weights1000)).loc[:,'2017-01-03':].dropna(how='all').mean()-1).rename('中证1000').dropna()
```

```python
Index_ENP_fn = pd.concat([
    pre_raise50.rename('50'),
    pre_raise300.rename('300'),
    pre_raise500.rename('500'),
    pre_raise1000.rename('1000')
],axis=1).shift(1).dropna().T
```

```python
ENPchange = Index_ENP_fn.pct_change(axis=1).T.rolling(60).mean().T
```

## 评估

### cl-cl

```python
# 分组收益率
gr_ENPchange = xy.cal_rate_group(
    ENPchange.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    group_num=4,
    masks=1
)
gr_ENPchange.cumsum().plot()
```

```python
# 单因子强弱
ENPchange_select, ENPchange_rets = xy.single_Factor_backtest(
    ENPchange.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    _side=1,
    fee=0.000,
    name='ENPchange'
)
```

```python
# 因子截面IC趋势
rankIC = xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False).corrwith(ENPchange.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False))
rankIC.rolling(60).mean().plot(figsize=(15,8),grid=1)
rankIC.mean()
```

### cl_op

```python
# 分组收益率
gr_ENPchange = xy.cal_rate_group(
    ENPchange.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    group_num=4,
    masks=1
)
gr_ENPchange.cumsum().plot()
```

```python
# 单因子强弱
ENPchange_select, ENPchange_rets = xy.single_Factor_backtest(
    ENPchange.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    _side=1,
    fee=0.000,
    name='ENPchange'
)
```

```python
# 因子截面IC趋势
rankIC = xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False).corrwith(ENPchange.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False))
rankIC.rolling(60).mean().plot(figsize=(15,8),grid=1)
rankIC.mean()
```
