```python
import pandas as pd
import numpy as np
from scipy.io import loadmat as loadmat
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['axes.unicode_minus'] = False

# Suppress warnings`
warnings.filterwarnings("ignore")
```

```python
import json
from datetime import datetime
import xy
```

```python
def read_json_from_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data
```

```python
def timestamp_to_date_string(timestamp_str):
    # Convert the timestamp string to an integer (assuming it's in milliseconds)
    timestamp = int(timestamp_str)

    # Convert the Unix timestamp to a datetime object
    # Here, we divide by 1000 to convert milliseconds to seconds
    datetime_obj = datetime.fromtimestamp(timestamp / 1000)

    # Format the datetime object as a string in your desired format
    date_string = datetime_obj.strftime('%Y-%m-%d')  # Change the format as needed

    return date_string

def trans_keys(data_dict):
    date_lst = []
    for timestamp_str in data_dict.keys():
        date_str = timestamp_to_date_string(timestamp_str)
        date_lst.append(date_str)
    
    Ser = pd.Series(dict(zip(date_lst,data_dict.values())))
    
    return Ser
```

## 增长预期

```python
file_path = 'BondCG7_10.txt'
data_dict = read_json_from_file(file_path)
CGBond7_10 = trans_keys(data_dict)
```

```python
CUI = pd.read_csv('CUI.csv',index_col=0)
HSI = pd.read_csv('HSI.csv',index_col=0)
CRBI = pd.read_csv('CRBI.csv',index_col=0)
```

```python
HSI_yoy = HSI.rolling(20).mean().pct_change(252).iloc[:,0]
CRBI_yoy = CRBI.rolling(20).mean().pct_change(252).iloc[:,0]
CUI_yoy = CUI.rolling(20).mean().pct_change(252).iloc[:,0]
CGBond7_10_yoy = (CGBond7_10.rolling(20).mean().pct_change(252))

growth_yoy = pd.concat([
    HSI_yoy.rename('HSI_yoy'),
    CRBI_yoy.rename('CRBI_yoy'),
    CUI_yoy.rename('CUI_yoy'),
    CGBond7_10_yoy.rename('CGBond7_10_yoy'),
],axis=1)
```

```python
weight_HSI = 1/HSI.rolling(20).mean().pct_change(252).rolling(20).std().iloc[:,0]
weight_CRBI = 1/CRBI.rolling(20).mean().pct_change(252).rolling(20).std().iloc[:,0]
weight_CUI = 1/CUI.rolling(20).mean().pct_change(252).rolling(20).std().iloc[:,0]
weight_CGBond7_10 = 1/CGBond7_10.rolling(20).mean().pct_change(20).rolling(252).std()

growth_weight = pd.concat([
    weight_HSI.rename('HSI'),
    weight_CRBI.rename('CRBI'),
    weight_CUI.rename('CUI'),
    weight_CGBond7_10.rename('CGBond7_10'),
],axis=1)
```

```python
def Singlation(x):
    return x / x.sum()
```

```python
growth_weight.dropna().T.apply(Singlation)
```

```python
HSI_yod = HSI.rolling(20).mean().pct_change(1).iloc[:,0]
CRBI_yod = CRBI.rolling(20).mean().pct_change(1).iloc[:,0]
CUI_yod = CUI.rolling(20).mean().pct_change(1).iloc[:,0]
CGBond7_10_yod = -CGBond7_10.rolling(20).mean().pct_change(1)

growth_yod = pd.concat([
    HSI_yod.rename('HSI'),
    CRBI_yod.rename('CRBI'),
    CUI_yod.rename('CUI'),
    CGBond7_10_yod.rename('CGBond7_10'),
],axis=1)
```

```python
growth = ((growth_yod.dropna().T * growth_weight.dropna().T.apply(Singlation)).sum().T + 1).cumprod().loc['2017-03-15':'2023-07-04'].pct_change(252).dropna()
```

```python
growth
```

```python
pd.concat([
    growth.loc[(growth.index & xy.Index_rets.loc[['300'],'2018-04-13':'2023-06-01'].T.cumsum().index)].rename('growth_yoy'),
    xy.Index_rets.loc[['1000'],'2018-04-13':'2023-06-01'].T.cumsum().loc[(growth.index & xy.Index_rets.loc[['1000'],'2018-04-13':'2023-06-01'].T.cumsum().index)]
],axis=1).plot()
```
