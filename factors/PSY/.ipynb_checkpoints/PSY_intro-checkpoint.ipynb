# $因子描述：心理线指标$

***N日内上涨的天数***  

## 计算

```python
st_PSY10 = (xy.rets > 0).T.rolling(10).sum().T.shift(1,axis=1)
st_PSY12 = (xy.rets > 0).T.rolling(12).sum().T.shift(1,axis=1)
st_PSY15 = (xy.rets > 0).T.rolling(15).sum().T.shift(1,axis=1)
st_PSY20 = (xy.rets > 0).T.rolling(20).sum().T.shift(1,axis=1)

mean_st_PSY = (st_PSY10 + st_PSY12 + st_PSY15 + st_PSY20)/3
```

```python
st_PSY = xy.read_Factor(xy.trd_dates,'st_PSY',df=mean_st_PSY)
```

## 评估

### cl-cl

```python
# 分组收益率
gr_st_PSY = xy.cal_rate_group(
    st_PSY.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    group_num=4,
    masks=1
)
gr_st_PSY.cumsum().plot()
```

```python
# 单因子强弱
st_PSY_select, st_PSY_rets = xy.single_Factor_backtest(
    st_PSY.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    _side=1,
    fee=0.000,
    name='st_PSY'
)
```

```python
# 因子截面IC趋势
rankIC = xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False).corrwith(st_PSY.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False))
rankIC.rolling(60).mean().plot(figsize=(15,8),grid=1)
rankIC.mean()
```

### cl_op

```python
# 分组收益率
gr_st_PSY = xy.cal_rate_group(
    st_PSY.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    group_num=4,
    masks=1
)
gr_st_PSY.cumsum().plot()
```

```python
# 单因子强弱
st_PSY_select, st_PSY_rets = xy.single_Factor_backtest(
    st_PSY.T.rolling(1).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    _side=1,
    fee=0.000,
    name='st_PSY'
)
```

```python
# 因子截面IC趋势
rankIC = xy.bonus_open_cl_rets.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False).corrwith(st_PSY.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False))
rankIC.rolling(60).mean().plot(figsize=(15,8),grid=1)
rankIC.mean()
```
