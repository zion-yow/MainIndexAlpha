# $因子描述：RCP$

***计算标的每天前230分钟当中的成交量峰值个数，识别短期内多次放量的行为。***
即日内出现“放量”现象越多的标的可能有更多的趋势交易者或知情交易者参与，进而有更高的概率在未来获得更高收益。

1. 数据：标的分钟行情（**对于指数因子，此处采用指数ETF分钟行情**）  
2. 定义每日成交量波峰，同时满足下列两个条件：  
    条件1：分钟成交量 - mean > k * std  
    条件2：**前一个分钟成交量与条件1互斥**   
3. 数每日波峰的数量

$参考报告：《20201012-长江证券-基础因子研究（十四）：高频因子（九），高频波动中的时间序列信息》$

## 计算

```python
def func2(arr):
    return np.nanmedian(np.argwhere(arr))
```

## 过度自信

```python
def index_CP_daily(arr, k):
    try:
        arr = arr.dropna()
        price = xy.rolling_window(arr,20)
        mean_price = np.tile(arr.rolling(20).mean().dropna(),(20,1)).T
        std_price = np.tile(arr.rolling(20).std().dropna(),(20,1)).T

        bool_l_price = price<(mean_price-k*std_price)
        bool_h_price = price>(mean_price+k*std_price)

        sub = (np.apply_along_axis(func2,1,bool_l_price) - np.apply_along_axis(func2,1,bool_h_price))

        # display(sub)
        CP_intraday = pd.Series(sub, arr.index[19:])
        # CP_intraday.index = CP_intraday.index.map(lambda x: x[:10])

        return CP_intraday
    except Exception as e:
        display(e)
        return pd.Series(np.nan, arr.index[19:])
```

```python
def index_CP_intraday(arr, k):
    price = xy.rolling_window(arr.dropna(),240)[::240]
    mean_price = np.tile(arr.dropna().rolling(240).mean().dropna()[::240],(240,1)).T
    std_price = np.tile(arr.dropna().rolling(240).std().dropna()[::240],(240,1)).T

    bool_l_price = price<(mean_price-k*std_price)
    bool_h_price = price>(mean_price+k*std_price)
    
    CP_intraday = pd.Series((np.apply_along_axis(func2,1,bool_l_price) - np.apply_along_axis(func2,1,bool_h_price)),arr[::240].index)
    CP_intraday.index = CP_intraday.index.map(lambda x: x[:10])
    
    return CP_intraday
```

```python
iCP_intraday50 = index_CP_intraday(xy.IndexMin50jq.close.dropna(), 0.8)
iCP_intraday300 = index_CP_intraday(xy.IndexMin300jq.close.dropna(), 0.8)
iCP_intraday500 = index_CP_intraday(xy.IndexMin500jq.close.dropna(), 0.8)
iCP_intraday1000 = index_CP_intraday(xy.IndexMin1000jq.close.dropna(), 0.8)
# 参数平滑
iCP_intraday4 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(4).mean().shift(1)

iCP_intraday5 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(5).mean().shift(1)

iCP_intraday6 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(5).mean().shift(1)

iCP_intraday = (iCP_intraday4 + iCP_intraday5 + iCP_intraday6) / 3
```

```python
idn = '1000'
def cal_rcp(idn, intraday,iCP_intraday, win):
    RCP = {}
    for i, d in tqdm(enumerate(intraday_ret.columns[:-2])):
        x = intraday_ret.loc[idn, :d].iloc[-win:]
        y = iCP_intraday.dropna().T.loc[idn, :d].iloc[-win:]
        if len(y) < win :
            continue
        else:
            X = np.array([x,np.ones(win,)]).T
            r = (np.linalg.lstsq(X, y,rcond=None)[1]/win)[0]**0.5
            RCP[y.index[-1]] = r
    return pd.Series(RCP).shift().rename(idn)
        # break
```

```python
rcp1000 = cal_rcp('1000', intraday_ret, iCP_intraday, 30)
rcp300 = cal_rcp('300', intraday_ret, iCP_intraday, 30)
rcp500 = cal_rcp('500', intraday_ret, iCP_intraday, 30)
rcp50 = cal_rcp('50', intraday_ret, iCP_intraday, 30)
```

```python
idx_RCP = pd.concat([
    rcp50.rename('50'),
    rcp300.rename('300'),
    rcp500.rename('500'),
    rcp1000.rename('1000'),
],axis=1)
```
