```python
# 更新 A股日频行情
def get_stk_panel(
    code:str,
    start_date:str,
    end_date:str,
)->dict:
    
    rs = bs.query_history_k_data_plus(code,
        "date,open,high,low,close,preclose,volume,amount,adjustflag,turn,tradestatus,pctChg,isST",
        start_date=start_date, end_date=end_date,
        frequency="d", adjustflag="2")# 后复权

    
    data_list = []
    while (rs.error_code == '0') & rs.next():
        # 获取一条记录，将记录合并
        data_list.append(rs.get_row_data())
        
    result = pd.DataFrame(data_list, columns=rs.fields)[['date', 'open', 'high', 'low', 'close', 'preclose', 'volume', 'amount', 'turn', 'tradestatus', 'pctChg', 'isST']]
    
    return result.set_index('date').to_dict()
```

```python
# 个股收益率
rets = closes.pct_change(axis=1)

# 指数收盘点位
cl = pd.read_csv('..\index_close.csv',index_col=0).T
cl.index = ['50','300','500','1000']
cl.columns = [str(i)[:4] + '-' + str(i)[4:6] + '-' + str(i)[6:] for i in cl.columns]


# 指数开盘点位
op = pd.read_csv('..\index_open.csv',index_col=0).T
op.index = ['50','300','500','1000']
op.columns = [str(i)[:4] + '-' + str(i)[4:6] + '-' + str(i)[6:] for i in op.columns]

# 指数前收盘点位
pcl = pd.read_csv('..\index_preclose.csv',index_col=0).T
pcl.index = ['50','300','500','1000']
pcl.columns = [str(i)[:4] + '-' + str(i)[4:6] + '-' + str(i)[6:] for i in pcl.columns]

# 指数分红
div_exd = pd.read_csv('..\index_div_exd.csv',index_col=0)
div_exd.index = ['50','300','500','1000']
exop = (op + div_exd).dropna(axis=1)

# 日内实体涨幅
bar_ret = cl/exop- 1

# 交易日
trd_dates = sorted(list(set(rets.columns[1:])&set(weights50.columns)))

# 指数收益率
Index_rets = cl.pct_change(axis=1)
Index_rets = Index_rets.T.dropna().where(~np.isinf(Index_rets.T.dropna()),np.nan).T.loc[:,'2016-01-04':]

# 除权前隔夜收益率
ex_open_cl_ret = exop /  cl.shift(1,axis=1)
ex_open_cl_rets = ex_open_cl_ret - 1
```
