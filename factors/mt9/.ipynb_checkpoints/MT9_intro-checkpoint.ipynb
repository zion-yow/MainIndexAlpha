# $因子描述：MT （Magic Turn9）$

```python
def cal_mt9down(x):
    td = 0
    tail = 0
    
    while True:
        if (x[tail-1] < x[tail-5]):# and td >=-9:
            tail -= 1
            td -= 1
        else:
            break
    return td

def cal_mt9up(x):
    td = 0
    tail = 0
    
    while True:
        if x[tail-1] > x[tail-5]:# and td <=9:
            tail -= 1
            td += 1
        else:
            break
    return td
```

```python
def main_mt9func_fix_period(
    sig_HD,
    opening_dates,
    ret_day_hour,
    _profit_shrd_day_hour,
    _fee,
    _period
):
    pnl_dct_day_hour = {}
    profit_shrd_day_hour = _profit_shrd_day_hour
    # loss_shrd_day_hour = -0.03
    # max_pnl_dct_day_hour = {}
    fee_series = {}
    fee = _fee
    detail = {}

    for i, open_date in tqdm(enumerate(opening_dates)):
        sig = sig_HD.loc[open_date]
        ret_series = ret_day_hour.loc[open_date:][1:]
        period = 0
        # print(open_date,sig)
        for r,_sig in zip(ret_series,sig_HD.loc[open_date:][:-1]):
            period += 1
            current_date = str(sig_HD.loc[open_date:][:period+1].index[-1])
            CR = ret_day_hour.loc[open_date:][1:period+1].sum()
            # 持有当前头寸计算浮盈
            if (_sig != -sig) & (period < _period):
                # 未达到止盈止损条件
                if abs(CR) < profit_shrd_day_hour:
                    pnl_dct_day_hour[current_date] = r
                    # print(sig,_sig,period,open_date,current_date,CR)
                    detail[current_date] = {
                        'open_date':open_date,
                        'current_date':current_date,
                        'sig':sig,
                        '_sig':_sig,
                        'period':period,
                        'r':r,
                        'CR':CR,
                    }
                    continue

            else:
                pnl_dct_day_hour[current_date] = r
                detail[current_date] = {
                        'open_date':open_date,
                        'current_date':current_date,
                        'sig':sig,
                        '_sig':_sig,
                        'period':period,
                        'r':r,
                        'CR':CR,
                    }
                # print(sig,_sig,period,open_date,current_date,CR)
                break

        fee_series[current_date] = -_fee
    print(pd.Series(fee_series).sum())
    return pd.Series(pnl_dct_day_hour,ret_day_hour.index).fillna(0), pd.Series(fee_series,ret_day_hour.index).fillna(0),detail
```

```python
def main_mt9func_fix_period(
    sig_HD,
    opening_dates,
    ret_day_hour,
    _profit_shrd_day_hour,
    _fee,
    _period
):
    pnl_dct_day_hour = {}
    profit_shrd_day_hour = _profit_shrd_day_hour
    fee_series = {}
    fee = _fee
    detail = {}

    for i, open_date in tqdm(enumerate(opening_dates)):
        sig = sig_HD.loc[open_date]
        ret_series = ret_day_hour.loc[open_date:]
        period = 0

        for r,_sig in zip(ret_series,sig_HD.loc[open_date:][:-1]):
            
            period += 1
            current_date = str(sig_HD.loc[open_date:][:period+1].index[-1])
            CR = ret_day_hour.loc[open_date:][:period].sum()
            # 持有当前头寸计算浮盈
            if (_sig != -sig) & (period < _period):
                # 未达到止盈止损条件
                if abs(CR) < profit_shrd_day_hour:
                    pnl_dct_day_hour[current_date] = r
                    # print(sig,_sig,period,open_date,current_date,CR)
                    detail[current_date] = {
                        'open_date':open_date,
                        'current_date':current_date,
                        'sig':sig,
                        '_sig':_sig,
                        'period':period,
                        'r':r,
                        'CR':CR,
                    }
                    continue

            else:
                pnl_dct_day_hour[current_date] = r
                detail[current_date] = {
                        'open_date':open_date,
                        'current_date':current_date,
                        'sig':sig,
                        '_sig':_sig,
                        'period':period,
                        'r':r,
                        'CR':CR,
                    }
                # print(sig,_sig,period,open_date,current_date,CR)
                break

        fee_series[current_date] = -_fee
    print(pd.Series(fee_series).sum())
    return pd.Series(pnl_dct_day_hour,ret_day_hour.index).fillna(0), pd.Series(fee_series,ret_day_hour.index).fillna(0),detail
```

```python
def cal_upshadow(index_ret, high, low, cl, op):
    upShadow = pd.Series(np.nan,index = index_ret.index)
    upShadow.loc[index_ret[index_ret>0].index] = (high - cl)/cl
    upShadow.loc[index_ret[index_ret<=0].index] = (high - op)/op
    
    return upShadow

def cal_downshadow(index_ret, high, low, cl, op):
    downShadow = pd.Series(np.nan,index = index_ret.index)
    downShadow.loc[index_ret[index_ret>0].index] = (-low + op)/op
    downShadow.loc[index_ret[index_ret<=0].index] = (-low + cl)/cl
    
    return downShadow
```

```python
def ouput_res(
    new_res_detail,
    ret_day_shadow,
    side
):
    pnl_max_ret = pd.Series((new_res_detail.groupby('open_date').HCR.max()-0.001),ret_day_shadow.index).fillna(0)
    pnl_min_ret = pd.Series((new_res_detail.groupby('open_date').LCR.min()-0.001),ret_day_shadow.index).fillna(0)
    
    max_R = pnl_max_ret.copy()
    max_R.index = max_R.index.map(lambda x:str(x)[:10])

    res_max = bt_result(max_R.groupby(level=0).sum())
    p_result = pd.Series((new_res_detail.groupby('open_date').r.sum()-0.001),ret_day_shadow.index)
    
    # res_max['方向'] = side
    res_max[f'持有到期胜率_{side}'] = p_result.dropna()[p_result.dropna()>0].shape[0]/p_result.dropna()[p_result.dropna()!=0].shape[0]
    res_max[f'持有到期收益均值_{side}'] = pd.Series((new_res_detail.groupby('open_date').r.sum()-0.001),ret_day_shadow.index).mean()
    res_max[f'最优收益均值_{side}'] = pd.Series((new_res_detail.groupby('open_date').HCR.max()-0.001),ret_day_shadow.index).mean()
    res_max[f'最优收益持仓时长_{side}'] = pd.Series((new_res_detail.groupby('open_date').HCR.apply(np.argmax)+1),ret_day_shadow.index).mean()
    res_max[f'持有平均时长_{side}'] = pd.Series((new_res_detail.groupby('open_date').period.max()),ret_day_shadow.index).mean()
    res_max[f'次数_{side}'] = new_res_detail.open_date.unique().shape[0]

    min_R = pnl_min_ret.copy()
    min_R.index = min_R.index.map(lambda x:str(x)[:10])
    # res_max['最差胜率'] = bt_result(min_R.groupby(level=0).sum())['胜率']

    max_result = pd.Series((new_res_detail.groupby('open_date').HCR.apply(lambda x:x[-1])-0.001),ret_day_shadow.index)

    # res_max['最优胜率'] = max_result.dropna()[max_result.dropna()>0].shape[0]/max_result.dropna()[max_result.dropna()!=0].shape[0]
    res_max[f'观测时长（k线数量）'] = pl
    res_max[f'k线级别（单位：分钟）'] = f'{kr}'
    res_max[f'进出参考'] = f'[{ma_in_and_out[0]}]_[{ma_in_and_out[-1]}_{ma_in_and_out[0]}]'
    res_max[f'最优累计收益_{side}'] = res_max[f'最优收益均值_{side}'] * res_max[f'次数_{side}']
    res_max[f'持有到期累计收益_{side}'] = res_max[f'持有到期收益均值_{side}'] * res_max[f'次数_{side}']
    
    return res_max
```

```python
def get_sigs1(
    SIGS,
    beta,
    pre_times
):
    cumsum_shadow_lenth = 0

    log_dct = {}
    cumsum_shadow_lenth_dct_in = {}
    cumsum_shadow_lenth_dct_out = {}
    log_dct[SIGS.dropna().index[0]] = 0
    # log_dct[SIGS.dropna().index[1]] = 0
    buyable = 0
    SIGG = SIGS.to_dict()
    SIGG['b'] = {}
    
    for time, row in SIGS.dropna().iloc[1:,:].iterrows():
        # print(time, pre_times[time],log_dct[pre_times[time]])
        if log_dct[pre_times[time]] == 1:
            if row.close <= row.ma_out_series:
                log_dct[time] = 0
                buyable = 0
                SIGG['b'][time] = buyable
            else:
                log_dct[time] = 1
                SIGG['b'][time] = buyable
                
        elif log_dct[pre_times[time]] == -1:
            if row.close >= row.ma_out_series:
                log_dct[time] = 0
                buyable = 0
                SIGG['b'][time] = buyable
            else:
                log_dct[time] = -1
                SIGG['b'][time] = buyable

        else:
            if  (abs(row.mt) == 8) :
                if row.mt > 0:
                    buyable = 1
                    log_dct[time] = 0
                    SIGG['b'][time] = buyable
                else:
                    buyable = -1
                    log_dct[time] = 0
                    SIGG['b'][time] = buyable

            else :
                if buyable == 0:
                    log_dct[time] = 0
                    SIGG['b'][time] = buyable
                    continue
                else:
                    if (buyable == 1) & (row.close <= row.ma_in_series):
                        log_dct[time] = -1
                        buyable = 0
                        SIGG['b'][time] = buyable

                    elif (buyable == -1) & (row.close >= row.ma_in_series):
                        log_dct[time] = 1
                        buyable = 0
                        SIGG['b'][time] = buyable
                    else:
                        log_dct[time] = 0
                        SIGG['b'][time] = buyable
            
    return log_dct,cumsum_shadow_lenth_dct_in, cumsum_shadow_lenth_dct_out,SIGG
```

```python
def get_sigs(
    SIGS,
    beta,
    pre_times
):
    cumsum_shadow_lenth = 0

    log_dct = {}
    cumsum_shadow_lenth_dct_in = {}
    cumsum_shadow_lenth_dct_out = {}
    log_dct[SIGS.dropna().index[0]] = 0
    buyable = 0
    
    for time, row in SIGS.dropna().iloc[1:,:].iterrows():
# --------------------------- 达到 9 ------------------------------------
        if  (abs(row.mt) >= 8) | (abs(row.pre_1mt) >= 8):
            # 已有头寸的情况
            if log_dct[pre_times[time]] == 1:
                if row.close <= row.ma_out_series:
                    log_dct[time] = 0
                    cumsum_shadow_lenth_dct_out[time] = row.close - row.ma_out_series
                else:   
                    log_dct[time] = 1
                    cumsum_shadow_lenth_dct_out[time] = row.close - row.ma_out_series
                    
            elif log_dct[pre_times[time]] == -1:
                if row.close >= row.ma_out_series:
                    log_dct[time] = 0
                    cumsum_shadow_lenth_dct_out[time] = -(row.close - row.ma_out_series)
                else:   
                    log_dct[time] = -1
                    cumsum_shadow_lenth_dct_out[time] = -(row.close - row.ma_out_series)
            
            # 没有头寸的情况
            # 向下反转
            else:
                if row.mt > 0:
                    # cumsum_shadow_lenth = row.up
                    if row.close <= row.ma_in_series:
                        log_dct[time] = 1
                        cumsum_shadow_lenth_dct_in[time] =  row.ma_in_series - row.close
                        # cumsum_shadow_lenth = 0
                    else:
                        log_dct[time] = np.nan
                        cumsum_shadow_lenth_dct_in[time] = row.ma_in_series - row.close

                # 向上反转
                else:
                    # cumsum_shadow_lenth = row.down
                    if row.close >= row.ma_in_series:
                        log_dct[time] = -1
                        cumsum_shadow_lenth_dct_in[time] = row.close - row.ma_in_series
                        # cumsum_shadow_lenth = 0
                    else:
                        log_dct[time] = np.nan
                        cumsum_shadow_lenth_dct_in[time] = row.close - row.ma_in_series

# --------------------------- 未达到 9 ------------------------------------
        else:
            # 已有头寸的两种情况
            if log_dct[pre_times[time]] == 1:
                if row.close <= row.ma_out_series:
                    log_dct[time] = 0
                    cumsum_shadow_lenth_dct_out[time] = row.close - row.ma_in_series
                else:   
                    log_dct[time] = 1
                    cumsum_shadow_lenth_dct_out[time] = row.close - row.ma_in_series
                    
            elif log_dct[pre_times[time]] == -1:
                if row.close >= row.ma_out_series:
                    log_dct[time] = 0
                    cumsum_shadow_lenth_dct_out[time] = -(row.close - row.ma_in_series)
                else:   
                    log_dct[time] = -1
                    cumsum_shadow_lenth_dct_out[time] = -(row.close - row.ma_in_series)
            
            # 没有头寸的情况
            else:
                log_dct[time] = 0
                
            continue 
            
    return log_dct,cumsum_shadow_lenth_dct_in, cumsum_shadow_lenth_dct_out
```
