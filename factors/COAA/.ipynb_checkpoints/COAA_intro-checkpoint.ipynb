# $因子描述：量在价先（Correlation of Return And Amount）$

***选取过去二十日中，高波动时段内的平均收益、波动率。***  
错期后的价量相关系数，反应了日内放量后拉升的显著性。

1. 对20日同时刻成交量做标准化处理，并剔除收益率为0的k线
2. cora:计算日内分钟数据的成交量和收益率的相关系数；
3. cora_A:在计算cora前错期处理，成交量在收益率前；
4. cora_R:在计算cora前错期处理，收益率在成交量前；

$参考报告：《20200227-方正证券-量价关系的高频乐章》$

## 计算

```python
def adj(st_ser,min_times):
    m = np.nanmean(xy.rolling_window(xy.rolling_window(st_ser,240)[::240].T, 20),axis=2).T.reshape(-1,)
    std = np.nanstd(xy.rolling_window(xy.rolling_window(st_ser,240)[::240].T, 20),axis=2).T.reshape(-1,)
    # print(m.shape)
    return pd.Series((st_ser[(20-1)*240:] - m)/std, index = min_times[(20-1)*240:])


def cal_COXA_st(min_index):
    # try:
    st0 = min_index.copy()
    min_times = st0.index
    pre_close = st0.close.shift(240)[239::240].values
    
    st0['preclose'] = np.tile(pre_close,(240,1)).T.reshape(-1,1)
    
    # adj
    # st0['volume'] = adj(st0.volume,min_times)
    # display(st0)
    st0['close'] = st0['close'][(20-1)*240:]
    st0['preclose']  = st0['preclose'][(20-1)*240:]
    
    # st0['ret'] = st0.close/st0.preclose - 1

    drop = 0

    win_vol = xy.rolling_window(st0.volume,220 - drop)[10::240]
    win_ret = abs(xy.rolling_window(st0.close,220 - drop)[10::240])
    win_close = xy.rolling_window(st0.close,220 - drop)[10::240]

    # 量先
    win_vol_fore = xy.rolling_window(st0.volume,219 - drop)[10::240]
    win_ret_fore = abs(xy.rolling_window(st0.close,219 - drop)[11::240])
    win_close_fore = xy.rolling_window(st0.close,219 - drop)[11::240]

    # Ret先
    win_vol_lag = xy.rolling_window(st0.volume,219 - drop)[11::240]
    win_ret_lag = abs(xy.rolling_window(st0.close,219 - drop)[10::240])
    win_close_lag = xy.rolling_window(st0.close,219 - drop)[10::240]

    cora = {}
    copa = {}
    cora_A = {}
    copa_A = {}
    cora_R = {}
    copa_P = {}
    
    times = min_times.map(lambda x : x[:10])
    times = times.unique()
    # display(win_ret_fore)
    for i,date in enumerate(times):

        # shape_fore = [~np.isnan(win_ret_fore[i]) * (~np.isnan(win_vol_fore[i]))]
        shape_fore = [~np.isnan(win_ret_fore[i]) * (~np.isnan(win_vol_fore[i])) * (win_ret_fore[i] != 0)]
        shape_fore = np.array(shape_fore).reshape(-1)
        
        if np.isnan(np.nanmean(win_ret_fore[i])):
            # print( ~any(win_ret_fore[i]))
            continue
        else:
            cora_A[date] = np.corrcoef(win_ret_fore[i]*shape_fore, win_vol_fore[i]*shape_fore)[0,1]

        shape_lag = [~np.isnan(win_ret_lag[i]) * (~np.isnan(win_vol_lag[i])) * (win_ret_lag[i] != 0)]
        shape_lag = np.array(shape_lag).reshape(-1)
        
        cora_R[date] = np.corrcoef(win_ret_lag[i]*shape_lag, win_vol_lag[i]*shape_lag)[0,1]

    return cora_A, cora_R
    
    # except Exception as e:
    #     return {}, {}
```

```python
cora_A50, cora_R50 = cal_COXA_st(xy.ETFMin50jq)
cora_A300, cora_R300 = cal_COXA_st(xy.ETFMin300jq)
cora_A500, cora_R500 = cal_COXA_st(xy.ETFMin500jq)
cora_A1000, cora_R1000 = cal_COXA_st(xy.ETFMin1000jq)
```

```python
COAR = pd.concat([
    pd.Series(cora_A50).rename('50'),
    pd.Series(cora_A300).rename('300'),
    pd.Series(cora_A500).rename('500'),
    pd.Series(cora_A1000).rename('1000')
          ],axis=1).shift(1).dropna(how='all')

CORA = pd.concat([
    pd.Series(cora_R50).rename('50'),
    pd.Series(cora_R300).rename('300'),
    pd.Series(cora_R500).rename('500'),
    pd.Series(cora_R1000).rename('1000')
          ],axis=1).shift(1).dropna(how='all')

COR = CORA+COAR
```

### cl_cl

```python
# 分组收益率
gr_COAR = xy.cal_rate_group(
    COR.rolling(20).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    group_num=4,
    masks=1
)
gr_COAR.dropna().cumsum().plot()
```

```python
# 单因子强弱
COAR_select, COAR_rets = xy.single_Factor_backtest(
    COAR.rolling(20).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':],
    xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':],
    _side=1,
    fee=0.000,
    name='COAR'
)
```

```python
# 因子截面IC趋势
rankIC = xy.Index_rets.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False).corrwith(COAR.rolling(20).mean().T.loc[['50', '300', '500', '1000'],'2016-02-28':].rank(ascending=False))
rankIC.rolling(60).mean().plot(figsize=(15,8),grid=1)
rankIC.mean()
```
