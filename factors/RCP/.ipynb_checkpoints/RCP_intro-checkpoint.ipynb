# $因子描述：CP$

***过度自信***  
衡量日内先"高开低走"程度。

1. 数据：标的分钟行情（**对于指数分钟行情**）  
2. 定义上行阶段和下行阶段：  
    上行：up = 分钟收盘价 - mean > k * std  
    下行：down = 分钟收盘价 + mean < k * std   
3. down的索引中位数 - up的索引中位数,得到当日因子值  

$参考报告：《20220915-东吴证券-金工专题报告:寻找特质波动率中的纯真信息,剔除跨期截面相关性的纯真波动率因子》$

## 计算

```python
def func2(arr):
    return np.nanmedian(np.argwhere(arr))
```

## 过度自信

```python
def index_CP_daily(arr, k):
    try:
        arr = arr.dropna()
        price = xy.rolling_window(arr,20)
        mean_price = np.tile(arr.rolling(20).mean().dropna(),(20,1)).T
        std_price = np.tile(arr.rolling(20).std().dropna(),(20,1)).T

        bool_l_price = price<(mean_price-k*std_price)
        bool_h_price = price>(mean_price+k*std_price)

        sub = (np.apply_along_axis(func2,1,bool_l_price) - np.apply_along_axis(func2,1,bool_h_price))

        # display(sub)
        CP_intraday = pd.Series(sub, arr.index[19:])
        # CP_intraday.index = CP_intraday.index.map(lambda x: x[:10])

        return CP_intraday
    except Exception as e:
        display(e)
        return pd.Series(np.nan, arr.index[19:])
```

```python
def index_CP_intraday(arr,v_arr,k):
    vol_ratio = xy.rolling_window(arr.dropna(),240)[::240][1:]/np.array(arr.dropna().iloc[239::240][1:]).reshape(-1,1)
    price = xy.rolling_window(arr.dropna(),240)[::240][1:]
    price = price * vol_ratio

    mean_price = np.tile(arr.dropna().rolling(240).mean().dropna()[::240][1:],(240,1)).T
    std_price = np.tile(arr.dropna().rolling(240).std().dropna()[::240][1:],(240,1)).T

    bool_l_price = price<(mean_price-k*std_price)
    bool_h_price = price>(mean_price+k*std_price)
    
    CP_intraday = pd.Series((np.apply_along_axis(func2,1,bool_l_price) - np.apply_along_axis(func2,1,bool_h_price)),arr[::240].index[1:])
    CP_intraday.index = CP_intraday.index.map(lambda x: x[:10])
    
    return CP_intraday
```

```python
iCP_intraday50 = index_CP_intraday(xy.IndexMin50jq.close.dropna(), xy.IndexMin50jq.money.dropna(), 0.9)
iCP_intraday300 = index_CP_intraday(xy.IndexMin300jq.close.dropna(), xy.IndexMin300jq.money.dropna(), 0.9)
iCP_intraday500 = index_CP_intraday(xy.IndexMin500jq.close.dropna(), xy.IndexMin500jq.money.dropna(), 0.9)
iCP_intraday1000 = index_CP_intraday(xy.IndexMin1000jq.close.dropna(), xy.IndexMin1000jq.money.dropna(), 0.9)
# 参数平滑
iCP_intraday1 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(1).mean().shift(1)

iCP_intraday3 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(3).mean().shift(1)

iCP_intraday4 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(4).mean().shift(1)

iCP_intraday5 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(5).mean().shift(1)

iCP_intraday6 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(6).mean().shift(1)

iCP_intraday7 = pd.concat([
    iCP_intraday50.rename('50'),
    iCP_intraday300.rename('300'),
    iCP_intraday500.rename('500'),
    iCP_intraday1000.rename('1000'),
],axis=1).rolling(7).mean().shift(1)

iCP_intraday = (iCP_intraday3 + iCP_intraday4 + iCP_intraday5 + iCP_intraday6 + iCP_intraday7) / 5
```
